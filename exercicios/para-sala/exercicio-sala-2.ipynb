```python
def processamento_df(df_titanic):
    #saber quantas linhas vai mostrar
    df_titanic.head(3)

    #criar um pequeno data frame
    baby_df = df_titanic.sample(100)

    #trazer valores aleatorios
    df_titanic = df_titanic.sort_values('Fare')

    #remorer as duplicadas de determinadas colunas como o 'drop' você remove o que voce quiser, linhas ou colunas
    df_titanic = df_titanic.drop(['SibSp', 'Parch'], axis = 'columns')

    #remover todas as linhas null, com o 'drop' você deleta tudo que é null
    df_titanic = df_titanic.dropna(subset = 'Fare')

    #redefine o indice (id) e colocar em na ordem certa
    df_titanic = df_titanic.reset_index()

    #traz as informações dos números
    metricas = df_titanic.describe()

    return baby_df, df_titanic, metricas

```

```python
#ler novamente o CSV
df_titanic = pd.read_csv('titanic.csv')
```

```python
baby_df, df_titanic, metricas = processamento_df(df_titanic)
```

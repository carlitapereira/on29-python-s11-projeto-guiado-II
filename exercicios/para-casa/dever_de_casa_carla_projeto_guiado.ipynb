```python

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('df_menor.csv')


g20_countries = ['Argentina', 'Australia', 'Brazil', 'Canada', 'China', 'France', 'Germany', 'India', 'Indonesia',
                 'Italy', 'Japan', 'Mexico', 'Russia', 'Saudi Arabia', 'South Africa', 'South Korea', 'Turkey',
                 'United Kingdom', 'United States']

df_g20_2017 = df[(df['Year'] == 2017) & (df['Entity'].isin(g20_countries))]


plt.figure(figsize=(10, 6))
plt.bar(df_g20_2017['Entity'], df_g20_2017['Schizophrenia (%)'])
plt.title('Prevalência de Esquizofrenia em 2017 (Países do G20)')
plt.xlabel('País')
plt.ylabel('Prevalência de Esquizofrenia (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Passo 1: Importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

# Passo 2: Ler o arquivo CSV em um DataFrame
df = pd.read_csv('df_menor.csv')

# Passo 3: Filtrar os dados para o ano de 2017 e para o país 'Brazil'
df_brazil_2017 = df[(df['Year'] == 2017) & (df['Entity'] == 'Brazil')]

# Passo 4: Selecionar apenas as colunas desejadas
df_brazil_2017 = df_brazil_2017[['Schizophrenia (%)', 'Bipolar disorder (%)', 'Eating disorders (%)', 'Anxiety disorders (%)', 'Drug use disorders (%)', 'Depression (%)', 'Alcohol use disorders (%)']]

# Passo 5: Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
df_brazil_2017.sum().plot(kind='bar', color='skyblue')
plt.title('Prevalência de Transtornos Mentais no Brasil em 2017')
plt.xlabel('Transtorno Mental')
plt.ylabel('Total (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python


df = pd.read_csv('df_menor.csv')


df_brazil = df[df['Entity'] == 'Brazil']


plt.figure(figsize=(15, 6))
plt.plot(df_brazil['Year'], df_brazil['Anxiety disorders (%)'], marker='o', color='skyblue', linestyle='-')
plt.title('Prevalência de Transtornos de Ansiedade no Brasil')
plt.xlabel('Ano')
plt.ylabel('Prevalência (%)')
plt.xticks(df_brazil['Year'])
plt.grid(True)
plt.tight_layout()
plt.show()
```
